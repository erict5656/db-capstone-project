```python
# Define the query to show all tables
show_tables_query = "SHOW tables"

# Execute the query
cursor.execute(show_tables_query)

# Fetch and display the results
results = cursor.fetchall()
print("Tables in the database:")
for table in results:
    print(table[0])
```

```python
# Define the query to join tables and apply the filter
query = """
SELECT Customers.FullName, Customers.ContactNumber, Orders.TotalCost
FROM Customers
JOIN Orders ON Customers.CustomerID = Orders.CustomerID
WHERE Orders.TotalCost > 60
"""

# Execute the query
cursor.execute(query)

# Fetch and display the results
results = cursor.fetchall()
print("Customers with orders greater than $60:")
for row in results:
    print(f"Full Name: {row[0]}, Contact Number: {row[1]}, Total Cost: ${row[2]}")
```
